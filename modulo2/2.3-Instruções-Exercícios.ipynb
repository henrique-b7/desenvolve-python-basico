### Q1

Faça um programa que armazene o valor de 20 reais de salário por hora de um trabalhador, e o número de 40 horas trabalhadas na semana. Com essas informações, calcule e imprima:
- Salário semanal bruto
- Valor descontado por semana INSS (10% do bruto)
- Valor descontado por semana pelo sindicato (5% do bruto)
- Salário semanal líquido (Bruto - Descontos)

```python
## Escreva e execute seu código aqui
salhs = 20
hssem = 40
salsem = salhs * hssem
print ("O salário semanal é R$", salsem)
inss = salsem * 10 / 100
print ("O desconto do INSS é R$", inss)
sind = salsem *5 / 100
print ("O desconto da contribuição sindical é R$", sind)
saliq = salsem - sind - inss
print ("O Salário Semanal Líquido é R$", saliq)
```

A fórmula para calcular a velocidade média ($v\_media$) é dada por $v\_media = \frac{distancia}{tempo}$ . Escreva um programa em Python para calcular a velocidade média necessária para completar a maratona em 3 horas.

**Calcule e imprima o resultado em metros/segundo**, ou seja, converta a distância de quilômetros para metros ($metro = \frac{km}{1000}$) e a velocidade de horas para segundos $(segundo = \frac{hora}{3600})$.

```python
## Escreva e execute seu código aqui
v_media = 42.195 / 3
print ("velocidade média necessária para completar a corrida:", v_media, "Km/H")
mps = v_media / 3.6
print ("velocidade média necessária para completar a corrida:", mps, "m/s")
```
